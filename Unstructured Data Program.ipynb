Citibike is the nationâ€™s largest bike share program, with approximately 12,000 bikes and 750 stations across Manhattan, Brooklyn, Queens, and Jersey City. This program is to facilite the bike rental experience by providing the amount of bikes available at each dock and locate stations where bikes are unavailable.

```python
''' 
	This program accesses the Citi Bike NYC data feed. This URL is for
       bike rentals.
	It then saves the station list to a database named bk_json and a collection called bikecoll.
	The mongod server must be running for this program to work!
'''

import urllib.request
import json
import pymongo
##open the url that accesses the json data
bike_url= "https://feeds.citibikenyc.com/stations/stations.json"
try:
	response = urllib.request.urlopen(bike_url)
except urllib.error.URLError as e:
    if hasattr(e, 'reason'):
        print('We failed to reach a server.')
        print('Reason: ', e.reason)
    elif hasattr(e, 'code'):
        print('The server couldn\'t fulfill the request.')
        print('Error code: ', e.code)
else:
	# the url request was successful - convert the response to a string
	json_string = response.read().decode('utf-8')
    	# the json package loads() converts the string to python dictionaries and lists
	bk_json = json.loads(json_string)
bk_json
bikelist = bk_json['stationBeanList']
bikelist
# Connection to Mongo DB
try:
   client=pymongo.MongoClient('localhost', 27017)
   print ("Connected successfully!!!")
except pymongo.errors.ConnectionFailure as e:
    print ("Could not connect to MongoDB: %s" % e )
bkdb=client.stations
bikecoll=bkdb.bikes
bikecoll.insert_many(bikelist)
print("Added", len(bikelist), "to bike collection in stations database")
###Grab back the collection of bikes from MongoDB
bike_docs=bikecoll.find()
###Loop through and print out location, available docks, total docks and total bikes for each station
for doc in bike_docs:
    station=doc['stationName']
    availabledocks=doc['availableDocks']
    totaldocks=doc['totalDocks']
    totalbikes=doc['availableBikes']
    ##Create labels
    print('Station Location:',station,'     Available Docks:',availabledocks,'   Total Docks:',totaldocks,'   Total Bikes:',totalbikes)
##Find stations where bike rentals are unavailable
results=bikecoll.find({'statusKey':{"$gt":1}})
for result in results:
    print(result)
```
